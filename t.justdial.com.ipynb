```python
import requests
from bs4 import BeautifulSoup 
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd
```

```python
driver = webdriver.Firefox()
driver.get('https://t.justdial.com/Jaipur/Heart-Doctors/nct-10080172?fl=undefined&ismap=undefined&searchCase=&stype=category_list')
time.sleep(5)
elm=driver.find_element_by_tag_name('html')
i=0
while i<10:
    try:
        elm.send_keys(Keys.END)
    except:
        pass
    i+=1
    print(i)
    time.sleep(3)
res=driver.execute_script("return document.documentElement.outerHTML")
```

```python
res
```

```python
soup=BeautifulSoup(res,'html.parser')
soup.prettify()
```

```python
address=[]
for i in soup.find_all('span',class_="font11 adrselpsis"):
    address.append(i.text)
print(address)  
```

```python
ratings=[]
for i in soup.find_all('span',class_="cmngrn font12"):
    ratings.append(i.text)
print(len(ratings)) 
```

```python
review=[]
for i in soup.find_all('span',class_="cmnrtngtext font10"):
    review.append(i.text.split()[0])
print(len(review))    
```

```python
add_inf=[]
for i in soup.find_all('div',class_="rsextra font10"):
    try:
        add_inf.append(i.find('span',class_="rsgrytxt").text)
    except:
        add_inf.append('None_avail')
    #timing.append(i.text)
print(len(add_inf)) 
```

```python
name=[]
for i in soup.find_all('div',class_="rsname font13"):
    nm=i.text.split()[1:]
    n=str()
    for i in nm:
        n=n+' '+i
    name.append(n)
print(len(name))    
```

```python
table=pd.DataFrame({
     "Name":name,
        "Address":address,
         "Ratings":ratings,
          "Reviews":review
        })

    
```

```python
table.to_csv("./file.csv", sep=',',index=False)
```
